## 1. 프로젝트 개요

   ### 1-1. 프로젝트 선정동기

#### - 국내 건설기업의 해외건설 현황<br>
현재 다수의 국내 건설기업들은 국내 건축 시장의 한계점을 극복하고, 글로벌 건설 리더로써의 역량 발휘를 위해 해외 수주 및 해외 건설에 많은 투자를 하고 있는 실정이다. 다수의 플랜트, 랜드마크, 초고층빌딩, 주거용 임대시설 등을 중동, 아시아, 북미, 중남미 지역에 건설하고 있다. <br>
아래의 표는 해외건설시장에서 국내건설기업이 차지하고 있는 점유율을 보여준다. <br> ENR(Engineering News-Record)에 따르면, 국내건설기업은 13년도 해외매출기준 세계시장 점유율 7.8%로 6위를 차지하고 있다. 또한, 13년도 경쟁력 순위에서는 세계 7위를 차지하고 있다. 세부 지표별로는 시공경쟁력 4위, 설계경쟁력 19위, 가격경쟁력 5위로 평가되었다.<br>
 <br>


#### - 세계건설 시장 규모 현황 및 전망 <br>
GCPOE(Global Construction Perspectives and Oxford Economics)에 따르면, 글로벌 건설 시장 규모는 2015년 약 9조 2,980억 달러에서 2023년 약 16조 6,812달러로 증가할 것으로 전망된다. 글로벌 건설 시장은 연평균 10.91%로 성장해왔으며, 성장치가 꾸준하고 성장 잠재력이 높기 때문에 국내 건설 기업들은 앞으로 이 분야에 더 많은 투자를 할 것으로 예상된다. 현재 대부분의 국내 건설 기업들은 중동지역의 플랜트 개발에 많은 투자를 하고 있는 실정이다.<br>

<br>

아래의 표는 전세계 건설시장을 6개의 권역(아시아, 동유럽, 중동&아프리카, 북아메리카, 남아메리카, 서유럽)으로 구분하고, 각 권역의 2015년, 2023년 건설시장 규모를 수치화하였다. 자세히 살펴보면, 아시아 지역(17개 국가)의 2015년 건설시장 규모는 약 4조 8148억 달러이며, 2023년에는 약 9조 2588억 달러의 시장을 형성할 것으로 전망되고, 2015년 대비 약 192%의 성장률을 보인다. 두 번째로 규모가 큰 서유럽의 경우를 살펴보면, 2015년 1조 90211억 달러의 규모, 2023년 3조 1552억 달러의 규모, 167%의 성장률을 보인다.<br> 
규모와 성장률 비교해봤을 때, 확실히 아시아 권역이 가장 큰 규모와 성장률을 차지하는 것을 알 수 있고, 이에 따라 국내건설기업의 해외건설 사업을 아시아 지역에 집중한다면, 많은 성과를 달성할 수 있다고 판단된다. <br>


<br>

#### - 동남아시아 국가 도시화 및 주거 현황 <br>
동남아시아는 전 세계에서 도시화가 가장 빠르게 진행되고 있는 지역 중 하나다. 도시화율(전체인구 대비 도시인구 비율)은 1980년대 20%대에서 2013년 50%대 수준으로 증가하였고, 2050년 경에는 동남아시아 증가 인구 2억명 중 70% 이상이 도시인구의 증가에서 비롯될 것으로 예상된다. 아래 표는 동남아시아 국가 도시화의 상세 정보를 제공한다. 표에서 알 수 있듯이, 동남아시아는 2000년부터 2010년 동안, 도시면적이 3만 1,130.1km2 에서 3만 5,785.4km2로 약 4,655.3km2 증가하였는데, 이는 서울 면적의 약 7.7배에 달하는 엄청난 크기다. 또한, 동남아 도시인구도 2000년 대비 1억 6,560만명에서 2010년 2억 9,210만명으로 약 76.4% 급증한 것으로 나타난다. 따라서 도시화와 인구 증가에 따른 주거 정책 확보와 안정적인 주택 공급이 시급한 것으로 판단된다.<br>


<br>
<br>
위의 데이터 및 현황들을 종합해 봤을 때, 현재 국내 건설기업과 국내 건설분야는 해외 건설에 더 많은 초점과 사업의 중점을 두는 것이 적절하다고 판단된다. 특히, 성장가능성이 다른 대륙에 비해 월등히 높은 동남아시아 건설시장 중심으로 이를 실현한다면, 많은 경제적 및 사회적 효과를 이룰 것이라고 판단된다.

#### - 경제성, 수익성 분석 <br>
   각 국가별 공사비 산출<br><br>
   경제지표를 통한 분석 <br><br>
   1. 회수기간법 (PP; Payback Period Method)<br>
    : 투자대상으로부터 예상되는 미래의 현금흐름이 투자에 소요되는 모든 비용을 회수하는데 소요되는 기간의 장단으로써, 투자안을 평가하는 방법이다.
    <br><br>
   2. 회계적 이익률법 (ARR; Accounting Rate of Return Method)<br>
    : 회계적 이익률(= 연평균순이익/연평균투자액)을 기업에서 정한 목표 이익률과 비교하여 투자 의사결정을 하는 방법이다.
    <br><br>
   3. 내부 수익률법 (IRR: Internal Rate of Return Method)<br>
    : 현재 및 미래의 현금유출과 현금유입의 현재가치를 동일하게 하는 할인율을 구하여, 그 값과 목표수익률 및 이자율을 비교하여 투자할 가치가 있는지 평가하는 방법이다.
    <br><br>
   4. 순현재 가치법 (NPV: Net Present Value Method)<br>
    : 투자로부터 발생되는 현금유입 및 유출을 화폐의 시간적 가치를 고려하여 계산하는 것이며, 현금유입의 현재가치에서 현금유출의 현재가치를 차감한 금액이 (+) 이면, 경제성이 있으나, (-)이면 경제성이 없다고 판단하는 방법이다. (+)의 값이 크면 클수록 경제성이 높다.
    <br><br>
   5. 수익성 지수 (PI: Probability Index), BCR (Benefit/Cost Ratio)<br>
    : (총수입의 현가)/(총투자비의 현가) 값으로써, 그 지수가 1이상이면 수익성이 있는 것으로 판단한다.
    <br><br>

## 4. 분석을 위한 데이터의 가공

### 1. 각 국가별 공사비 산출을 위한 데이터의 가공 <br>
Turner & Towsend International construction market survey 2019.pdf 자료를 활용한 각국의 공사비 관련 세부 공사비 내역에 대한 데이터 획득 및 가공 <br><br>
데이터 위치 : ./data/Turner Townsend International Construction Market Survey 2019.pdf <br>


```python
import csv

with open('./data/Hochimin_USD.csv' , 'r') as fileRead:
    Hochimin_semi = []
    for hoc in fileRead:
        hoc2 = str(hoc.strip('\n'))
        Hochimin_semi.append(hoc2.split(','))
        
del Hochimin_semi[1] # 필요없는 것 삭제
Hochimin = Hochimin_semi[0]
        
#print("Hochimin data : ", Hochimin)


with open('./data/Jakarta_USD.csv' , 'r') as fileRead2:
    Jakarta_semi = []
    for jak in fileRead2:
        jak2 = str(jak.strip('\n'))
        Jakarta_semi.append(jak2.split(','))
        
del Jakarta_semi[1] # 필요없는 것 삭제
Jakarta = Jakarta_semi[0]
        
#print("Jakarta data : ", Jakarta)


with open('./data/Kuala_Lumpur_USD.csv' , 'r') as fileRead3:
    Kuala_semi = []
    for kul in fileRead3:
        kul2 = str(kul.strip('\n'))
        Kuala_semi.append(kul2.split(','))
        
del Kuala_semi[1] # 필요없는 것 삭제
Kuala = Kuala_semi[0]
        
#print("Kuala Lumpur data : ", Kuala)


with open('./data/Singapore_USD.csv' , 'r') as fileRead4:
    Singapore_semi = []
    for sin in fileRead4:
        sin2 = str(sin.strip('\n'))
        Singapore_semi.append(sin2.split(','))
        
del Singapore_semi[1] # 필요없는 것 삭제
Singapore = Singapore_semi[0]
        
#print("Singapore data : ", Singapore)


Total_data = []

i = 0
while i != len(Singapore):
    list = []
    list.append(Hochimin[i])
    list.append(Jakarta[i])
    list.append(Kuala[i])
    list.append(Singapore[i])
    Total_data.append(list)
    i += 1
    
#print(Total_data)

with open('./data/total_construction_data.csv', 'w') as fileWrite:
    writer = csv.writer(fileWrite)
    writer.writerow(Total_data)
    

from pandas import Series, DataFrame
import pandas as pd

frame = DataFrame(Total_data, columns = ['Ho chi min', 'Jakarta', 'Kuala Lumpur', 'Singapore'])
frame
```
